```python
import matplotlib.pyplot as plt

domains = [
    "datepanchang.com", "buynowfromusa.com", "fini.net", "lozo.com",
    "newstetic.com", "triggerfish.se", "afairjudgement.com",
    "owlcreek.com", "radioterminal.ru", "junio.com"
]
servers_visited = [3, 3, 3, 3, 3, 3, 3, 3, 3, 3]
latencies = [0.8699, 0.4168, 0.6242, 0.4531, 0.4366, 0.5107, 0.5787, 0.7551, 0.5808, 0.7633]

plt.figure(figsize=(10, 5))
plt.bar(domains, latencies, color = 'plum')
plt.title("Domain Name vs Latency Time")
plt.xlabel("Domain Name")
plt.ylabel("Latency (seconds)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(domains, servers_visited, color='paleturquoise')
plt.xticks(rotation=45, ha='right')
plt.ylabel("Number of Servers Visited")
plt.title("Domain Name vs Number of Servers Visited")
plt.tight_layout()
plt.show()
```
